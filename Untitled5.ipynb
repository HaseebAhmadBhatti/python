```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 4*np.pi, 1000)
Y = np.sin(X)
```

```python
fig, ax = plt.subplots()
ax.plot(X, Y)
fig.show()
```

```python
X = np.random.uniform(0, 1, 100)
Y = np.random.uniform(0, 1, 100)
ax.scatter(X, Y)
```

```python
X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
ax.bar(X, Y)
```

```python
Z = np.random.uniform(0, 1, (8,8))
ax.imshow(Z)
```
